```python
import matplotlib.pyplot as plt

mean = np.mean(amplitudes_0)
std = np.std(amplitudes_0)
mean_d = np.mean(-pulser_amplitudes_0123.data['Amplitude_ch0'])
std_d = np.std(-pulser_amplitudes_0123.data['Amplitude_ch0'])

plt.hist(amplitudes_0, histtype='step', bins=30, label='manual')
plt.hist(-pulser_amplitudes_0123.data['Amplitude_ch0'], histtype='step', bins=30, label='digitizer')
plt.title('Amplitude Channel 0')
plt.xlabel('Amplitude [V]')
plt.ylabel('Events')
plt.text(0.75, 0.7, f'Mean M: {mean:.4f}\nStd M: {std:.4f}\nMean D: {mean_d:.4f}\nStd D: {std_d:.4f}', transform=plt.gca().transAxes, fontsize=10)
plt.legend(loc=2, prop={'size':10})
plt.show()
```
